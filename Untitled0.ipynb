```python
# prompt: Create a correlation Matrix  of all the variables assigning all string values a number based on the hierarchy

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
import scipy as sc
import warnings


# Download the dataset from the UCI Machine Learning Repository
!wget https://archive.ics.uci.edu/ml/machine-learning-databases/00222/bank.zip

# Unzip the downloaded file
!unzip bank.zip

# Load the dataset into a pandas DataFrame
bank_data = pd.read_csv('bank-full.csv', sep=';')

# Print the first few rows of the dataset
print(bank_data.head())

# Create a scatter plot of age vs balance, colored by job
plt.figure(figsize=(12, 6))
sb.scatterplot(x='age', y='balance', hue='job', data=bank_data)
plt.title('Relationship between Age, Job, and Balance')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.show()

# You can also create a pairplot to visualize relationships between multiple variables
# sb.pairplot(bank_data[['age', 'balance', 'job']], hue='job')
# plt.show()


# Function to assign numerical values to string variables based on hierarchy
def assign_numerical_values(df, column_name, hierarchy_order):
  mapping = {value: index for index, value in enumerate(hierarchy_order)}
  df[column_name] = df[column_name].map(mapping)
  return df

# Identify columns with string values
string_columns = bank_data.select_dtypes(include=['object']).columns

# Define hierarchy orders for string columns
# Replace these with your desired hierarchy for each column
hierarchy_orders = {
    'job': ['admin.', 'blue-collar', 'entrepreneur', 'housemaid', 'management', 'retired', 'self-employed', 'services', 'student', 'technician', 'unemployed', 'unknown'],
    'marital': ['married', 'single', 'divorced'],
    'education': ['primary', 'secondary', 'tertiary', 'unknown'],
    'default': ['no', 'yes'],
    'housing': ['no', 'yes'],
    'loan': ['no', 'yes'],
    'contact': ['cellular', 'telephone'],
    'month': ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec'],
    'poutcome': ['failure', 'success', 'unknown']
}


# Assign numerical values based on the defined hierarchy
for column_name in string_columns:
  if column_name in hierarchy_orders:
    bank_data = assign_numerical_values(bank_data, column_name, hierarchy_orders[column_name])

# Calculate the correlation matrix
correlation_matrix = bank_data.corr()

# Print the correlation matrix
print(correlation_matrix)

# Visualize the correlation matrix with a heatmap
plt.figure(figsize=(12, 10))
sb.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
